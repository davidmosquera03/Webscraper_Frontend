```python
import numpy as np
import requests
from bs4 import BeautifulSoup
import random
```

```python
url = "https://www.bbc.com/mundo"
response = requests.get(url)
# verificamos el exito de la conexion con la url
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()
    text = text.split()
    # obtenemos el texto y con split lo separamos como arreglo
else:
    print('Error al obtener la página web:', response.status_code)
```

```python
soup
```

```python
h1_count = 0
h2_count = 0
h3_count = 0
```

```python

h1_tags = soup.find_all('h1')
for h1 in h1_tags:
    print(h1.text)
    h1_count += 1
```

```python

h2_tags = soup.find_all('h2')
for h2 in h2_tags:
    print(h2.text)
    h2_count += 1
```

```python
h3_tags = soup.find_all('h3')
for h3 in h3_tags:
    print(h3.text)
    h3_count += 1
```

```python
print("Cantidad de h1:", h1_count)
print("Cantidad de h2:", h2_count)
print("Cantidad de h3:", h3_count)
```
